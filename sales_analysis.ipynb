```python
df.fillna(0, inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
df.info()
```

```python
print(df['QUANTITYORDERED'].min())
print(df['PRICEEACH'].min())
print(df['SALES'].min())
```

```python
q1 = df['SALES'].quantile(0.25)
q3 = df['SALES'].quantile(0.75)
iqr=q3-q1
upper_limit = q3 + 1.5 * iqr
```

```python
df = df[df['SALES'] <= upper_limit]
```

```python
df.head()
```

```python
df.describe()
```

```python
print(upper_limit)
```

```python
df['YEAR'] = df['ORDERDATE'].dt.year
df['MONTH'] = df['ORDERDATE'].dt.month
df['QUARTER'] = df['ORDERDATE'].dt.quarter
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

monthly_revenue = df.groupby(['YEAR', 'MONTH'])['SALES'].sum().reset_index()
monthly_revenue['DATE'] = pd.to_datetime(monthly_revenue[['YEAR', 'MONTH']].assign(DAY=1))

plt.figure(figsize=(14, 5))
#width=14,height=5
sns.lineplot(data=monthly_revenue, x='DATE', y='SALES', marker='o', color='steelblue')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
product_revenue = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(data=product_revenue, x='PRODUCTLINE', y='SALES', palette='Blues_d')
plt.title('Revenue by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
country_revenue = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(14, 5))
sns.barplot(data=country_revenue, x='COUNTRY', y='SALES', palette='Greens_d')
plt.title('Revenue by Country')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
total_revenue   = df['SALES'].sum()
avg_order_value = df['SALES'].mean()
total_orders    = df['ORDERNUMBER'].nunique()
top_country     = country_revenue.iloc[0]['COUNTRY']

print(f"Total Revenue    : ${total_revenue:,.2f}")
print(f"Avg Order Value  : ${avg_order_value:,.2f}")
print(f"Total Orders     : {total_orders}")
print(f"Top Country      : {top_country}")
```

```python
status_counts = df['STATUS'].value_counts().reset_index()
status_counts.columns = ['STATUS', 'COUNT']

plt.figure(figsize=(8, 6))
sns.barplot(data=status_counts, x='STATUS', y='COUNT', palette='Oranges_d')
plt.title('Order Status Breakdown')
plt.xlabel('Status')
plt.ylabel('Number of Orders')
plt.yticks(range(0,2500,500))
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
quarterly_revenue = df.groupby(['YEAR', 'QUARTER'])['SALES'].sum().reset_index()
quarterly_revenue['YQ'] = 'Q' + quarterly_revenue['QUARTER'].astype(str) + '-' + quarterly_revenue['YEAR'].astype(str)

plt.figure(figsize=(12, 5))
sns.barplot(data=quarterly_revenue, x='YQ', y='SALES', palette='Purples_d')
plt.title('Quarterly Revenue Comparison')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
top_customers = df.groupby('CUSTOMERNAME')['SALES'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(12, 5))
sns.barplot(data=top_customers, x='SALES', y='CUSTOMERNAME', palette='Blues_d')
plt.title('Top 10 Customers by Revenue')
plt.xlabel('Total Sales')
plt.ylabel('Customer')
plt.tight_layout()
plt.show()
```

```python
monthly_revenue_sorted = monthly_revenue.sort_values('DATE')
monthly_revenue_sorted['MOM_GROWTH'] = monthly_revenue_sorted['SALES'].pct_change() * 100

plt.figure(figsize=(14, 5))
sns.lineplot(data=monthly_revenue_sorted, x='DATE', y='MOM_GROWTH', marker='o', color='tomato')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title('Month-over-Month Revenue Growth (%)')
plt.xlabel('Month')
plt.ylabel('Growth %')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
corr = df[['QUANTITYORDERED', 'PRICEEACH', 'SALES', 'MSRP']].corr()

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(corr, 
            annot=False,   # turn off seaborn annotations
            cmap='coolwarm', 
            linewidths=0.5,
            ax=ax)

# Manually write every value ourselves
for i in range(len(corr)):
    for j in range(len(corr.columns)):
        ax.text(j + 0.5, i + 0.5, 
                f'{corr.iloc[i, j]:.2f}',
                ha='center', va='center',
                color='black', fontsize=13, fontweight='bold')

plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
deal_counts = df['DEALSIZE'].value_counts().reset_index()
deal_counts.columns = ['DEALSIZE', 'COUNT']

plt.figure(figsize=(7, 5))
sns.barplot(data=deal_counts, x='DEALSIZE', y='COUNT', palette='coolwarm')
plt.title('Deal Size Distribution')
plt.xlabel('Deal Size')
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.show()
```

```python
deal_revenue = df.groupby('DEALSIZE')['SALES'].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(7, 5))
sns.barplot(data=deal_revenue, x='DEALSIZE', y='SALES', palette='magma')
plt.title('Revenue by Deal Size')
plt.xlabel('Deal Size')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```
